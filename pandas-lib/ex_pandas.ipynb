```
sr = pd.Series(index=range(6), data=[10, 11, 12, 13, 14, 15])
print(sr)
```

```
odd_even = sr.groupby(sr.values % 2) # grouped by series values
print(odd_even.get_group(0))
print(odd_even.get_group(1))
```

```
print(odd_even.max()) # print the maximum value in each group
```

```
# Group Selection in Datafame
dataF = pd.DataFrame({
  "col1": [0, 0, 1, 1],
  "col2": ["a", "b", "c", "d"],
  "col3": [2, 3, 4, 5]
})
dataF
```

```
grp = dataF.groupby("col1") # col1 wise group so we will get two group 0 and 1
grp.get_group(0)
```

```
grp.sum() # Only col3 will return cause it has numerical value
```

```
# Create new colum using eval in Dataframe
dataF["col1+col3"] = dataF.eval("col1 + col3")
dataF
```

```
grp = dataF.groupby(["col1", "col1+col3"])
grp.sum()
```

```
grp.sum().unstack()
```
