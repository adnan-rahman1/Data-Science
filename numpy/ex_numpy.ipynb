```
import numpy as np
```

```
x = np.array([1, 2, 3], dtype=np.float32)
x.itemsize
```

```
sinfunc = np.sin(np.array([1, 2, 3], dtype=np.float32))
print(sinfunc)
```

```
# Rank two tensor
x = np.array([[1, 2, 3], [4, 5, 6]])
print(x.shape)
print(x)
```

```
# Slicing in numpy
print(x[:, 0]) # All the rows and 0th column
print(x[:, 1]) # All the rows and 1st column
print(x[0, :]) # All the columns and 0th row
print(x[1, :]) # All the columns and 1st row
```

```
print(x[:, 0:2]) # All the rows and 1st & 2nd columns
```

```
# In numpy advance indexing create new copy of an array
m = np.ones((3, 3))
print(m)
```

```
n = m[:, [0, 1, 2, 2]] # Duplicate the last index and add another dimension
print(n.shape)
print(n)
```

```
# Here "n" has its own memeory 
# Changing any value of "m" does not affect "n"
m[0, 0] = 99
print(m)
print(n)
```

```
# Numpy array slicing create views not copy
m = np.array([4,5,6,7])
n = m[:2]
n[0] = 0
print(m)
print(n)
```

```
from numpy.lib.stride_tricks import as_strided
```

```
x = np.arange(16, dtype=np.int64)
print(x)
x.strides
```

```
y = as_strided(x, (7, 4), (16, 8))
print(y)
print(y.strides)
```

```
# Numpy matrix
A = np.matrix([[1, 2, 3], [4, 5, 6], [7,8, 9]]) # row = 3 & column = 3
x = np.matrix([[1],[0],[0]]) # row = 3 & column = 1
```

```
A @ x # same as A*x
```

```
A = np.ones((3, 3))
x = np.ones((3, 1))
print(np.matrix(A)*x)
```

```
# Broadcasting in numpy
X, Y = np.meshgrid(np.arange(4), np.arange(3))
```

```
print(X)
print(Y)
```

```
x = np.array([0, 1])
y = np.array([0, 1])
print(x)
print(y[:, None]) # convert to 2,1 matrix
```

```
z = y[:, None]
print(x+z) # broadcast x in axis = 0
```

```
z = np.array([0, 1, 2, 3])
z[:, None, None].shape
```

```
# Numpy mashed array
from numpy import ma
x = np.arange(10)
print(x)
y = ma.masked_array(x, x<5) # Creating mask does not copy the array
print(y)
print(y.shape)
```

```
x[-1] = 99
print(y)
```
