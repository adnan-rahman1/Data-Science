```
all_outcomes = { (i, j): i + j for i in range(1, 7) for j in range(1, 7) } # All the outcomes of throwing two dices
all_outcomes
```

```
sumUpSeven = { (i, j): i + j for i in range(1, 7) for j in range(1, 7) if i+j == 7 } # All the outcomes of throwing two dices sum up to 7
```

```
sumUpSeven
```

```
# We can do it in different way using defaultdict
from collections import defaultdict
dice = defaultdict(list) # defaultdict creates dictionaries with default values when it encounters a new key
```

```
for key, value in all_outcomes.items():
  dice[value].append(key)
```

```
dice
```

```
# Probability of all the outcomes of throwing two dices are
prob_2dice = { key: len(value) / 36.0 for key, value in dice.items() }
prob_2dice
```

```
import pandas as pd
d = pd.DataFrame(index=[ (i, j) for i in range(1, 7) for j in range(1, 7) ], columns = ["sm", "dice1", "dice2", "pd1", "pd2", "p"])
```

```
d.dice1 = [i[0] for i in d.index]
d.dice2 = [i[1] for i in d.index]
d.sm = list(map(sum, d.index))
d.head(10)
```

```
# Probability of unfair dice1
d.loc[d.dice1 <=3, "pd1"] = 1/9 
d.loc[d.dice1 > 3, "pd1"] = 2/9

# Probability of fair dice2
d.pd2 = 1/6

# joint Probability of dice1 and dice2
d.p = d.pd1 * d.pd2
d
```

```
# Probability mass function of sum of two dices
d.groupby("sm")["p"].sum()
```
